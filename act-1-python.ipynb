Este conjunto de datos proporciona información detallada sobre las sesiones de los visitantes de un sitio web de comercio electrónico. Se puede utilizar para predecir si un visitante realizará una compra o no.

## 3. Análisis exploratorio de datos (EDA)
Aquí realizaremos un análisis más profundo de las variables, tanto individuales como sus relaciones. Podemos hacer un resumen estadístico y visualizar algunos gráficos para comprender mejor los datos.

Resumen estadístico:


```python

# Visualización de la distribución de las variables numéricas
data.hist(bins=20, figsize=(20,15))
plt.show()
```

Relaciones entre variables:
Vamos a usar un mapa de calor para ver las correlaciones entre las variables continuas y entender qué factores podrían influir en el comportamiento de compra.

```python

# Excluir columnas no numéricas y binarias
numerical_data = data.drop(columns=['Month', 'VisitorType', 'Weekend', 'Revenue'])

# Renombrar columnas para mejor legibilidad
renamed_columns = {
    'Administrative': 'Admin',
    'Administrative_Duration': 'Admin_Duration',
    'Informational': 'Info',
    'Informational_Duration': 'Info_Duration',
    'ProductRelated': 'Product',
    'ProductRelated_Duration': 'Product_Duration',
    'BounceRates': 'Bounce_Rate',
    'ExitRates': 'Exit_Rate',
    'PageValues': 'Page_Values',
    'SpecialDay': 'Special_Day',
    'OperatingSystems': 'OS',
    'Browser': 'Browser',
    'Region': 'Region',
    'TrafficType': 'Traffic_Type'
}
numerical_data = numerical_data.rename(columns=renamed_columns)

# Calcular matriz de correlación
correlation_matrix = numerical_data.corr()

# Configuración del gráfico
plt.figure(figsize=(14, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)

# Agregar título y ajustar etiquetas
plt.title('Matriz de Correlación de Variables Numéricas', fontsize=16)
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()

```

En la matriz de correlación:

Relaciones fuertes positivas:

ProductRelated y ProductRelated_Duration tienen una correlación fuerte positiva (0.86). Esto indica que a mayor número de acciones relacionadas con productos, también aumenta la duración de dichas acciones.
BounceRates y ExitRates tienen una correlación muy fuerte (0.91). Esto sugiere que cuando la tasa de rebote es alta, también lo es la tasa de salida, lo cual es esperable, ya que ambos miden comportamientos similares de los usuarios al abandonar el sitio.
Relaciones moderadas:

Administrative y Administrative_Duration tienen una correlación positiva moderada (0.60). Esto significa que los usuarios que realizan más acciones administrativas tienden a pasar más tiempo en dichas acciones, pero no es una relación tan fuerte como con las acciones relacionadas con productos.
Informational y Informational_Duration también muestran una correlación moderada (0.62), lo que indica que a más acciones informativas, mayor duración, pero nuevamente no es una relación perfecta.
Relaciones negativas débiles:

Algunas variables como BounceRates y ProductRelated muestran una correlación negativa débil (-0.20). Esto podría sugerir que más acciones relacionadas con productos se asocian con una menor tasa de rebote, lo que podría indicar un mayor compromiso de los usuarios que interactúan con productos.
Variables independientes:

SpecialDay parece tener una relación casi nula con otras variables, lo cual es esperado ya que indica si un día es especial o no y probablemente no influya directamente en otras variables numéricas.
Relaciones no significativas:

Varias variables, como OS, Browser, Region, y TrafficType, tienen correlaciones muy cercanas a cero con la mayoría de las otras variables. Esto sugiere que estas variables no tienen una relación lineal significativa con el resto de los factores medidos en la matriz de correlación.
Conclusiones generales:
Acciones relacionadas con productos son un factor muy importante en términos de tiempo invertido por los usuarios.
BounceRates y ExitRates están altamente relacionadas, lo cual es natural dado que ambas miden el comportamiento de salida de los usuarios.
Variables como el sistema operativo y el navegador no parecen tener una relación significativa con el comportamiento de los usuarios en términos de las acciones o el tiempo que pasan en el sitio.


## 4. Visualización de resultados

```python
# Boxplot para la duración en páginas de producto, segmentado por intención de compra
plt.figure(figsize=(10, 6))
sns.boxplot(x='Revenue', y='BounceRates', data=data, hue='Revenue', palette='Set1', legend=False)
plt.title('boxplot 1: Distribución de la Tasa de Rebote por Intención de Compra')
plt.xlabel('Intención de Compra (Revenue)')
plt.ylabel('Tasa de Rebote')
plt.show()
```

### Distribución de la Tasa de Rebote por Intención de Compra

### Gráfica:
- El eje `x` muestra nuevamente la variable `Revenue` (intención de compra: `True` o `False`).
- El eje `y` muestra la `BounceRates`, es decir, la tasa de rebote (la proporción de sesiones en las que el usuario abandona la página sin realizar ninguna acción adicional).

### Interpretación:
- **Usuarios sin intención de compra (`False`)**:
  - La tasa de rebote es mayor, con la mediana más alta y un rango más amplio.
  - Hay muchos valores atípicos (outliers) hacia arriba, lo que sugiere que algunos usuarios sin intención de compra abandonan rápidamente la página.
- **Usuarios con intención de compra (`True`)**:
  - La tasa de rebote es más baja y la distribución está más concentrada en valores cercanos a cero.
  - Esto indica que los usuarios que tienen intención de compra tienden a explorar más las páginas, abandonando menos frecuentemente que los que no tienen intención de compra.

En resumen, los usuarios con intención de compra (`True`) tienden a tener tasas de rebote más bajas, lo que es un comportamiento esperado ya que están más comprometidos con el contenido de la página.



```python
# Boxplot para las tasas de rebote segmentado por intención de compra
plt.figure(figsize=(10, 6))
sns.boxplot(x='Revenue', y='ProductRelated_Duration', data=data, hue='Revenue', palette='Set2', legend=False)
plt.title('boxplot 2: Distribución del Tiempo en Páginas de Producto por Intención de Compra')
plt.xlabel('Intención de Compra (Revenue)')
plt.ylabel('Duración en Páginas de Producto (segundos)')
plt.show()
```

### Distribución del Tiempo en Páginas de Producto por Intención de Compra

### Gráfica:
- El eje `x` muestra la variable `Revenue`, que indica si hubo una intención de compra (`True`) o no (`False`).
- El eje `y` muestra la `ProductRelated_Duration`, que es el tiempo (en segundos) que los usuarios pasaron en páginas de productos.

### Interpretación:
- **Ambos grupos (`False` y `True`) tienen distribuciones muy similares**. La mayoría de los datos están cerca de cero, lo que indica que en promedio, los usuarios pasan poco tiempo en las páginas de productos, independientemente de si compran o no.
- **Los valores atípicos (outliers)**: Hay algunos puntos extremadamente altos que representan sesiones donde los usuarios pasaron una cantidad de tiempo considerablemente mayor en las páginas. Sin embargo, estos valores son raros.
- **Mediana y rango intercuartil**: Las cajas son muy pequeñas, lo que significa que el rango intercuartil (IQR, la diferencia entre el primer y tercer cuartil) es bajo. Esto indica que la mayoría de las sesiones tienen duraciones muy cortas.

En resumen, parece que el tiempo en las páginas de producto no varía mucho entre los usuarios que compran y los que no.


### Conclusión general:
- **boxplot 1** indica que los usuarios con intención de compra tienen una tasa de rebote considerablemente menor en comparación con aquellos sin intención de compra, lo que refleja un mayor compromiso.
- **boxplot 2** muestra que el tiempo en las páginas de producto es relativamente similar para todos los usuarios, con algunos casos excepcionales donde se pasa mucho más tiempo.
